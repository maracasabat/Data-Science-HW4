```python
import numpy as np
from scipy import integrate
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt
```

- Приймемо, що в нульовій координаті у нас швидкість 25 км/год, згідно з набором даних speed. Помістіть змінну time — вектор часу створений за допомогою np.linspace(всього 12 спостережень, від 0 до 11 годин)
- Виконайте вивід масиву time

```python
speed = [25, 35, 45, 30, 60, 120, 100, 100, 70, 75, 80, 65]
```

```python
time = np.linspace(0, 11, 12)
time
```

- Виконайте виведення графіка – значення точок швидкості ( plot або scatter). Встановіть розмір області (0, 11) та (0, 130). Задайте відображення сітки

```python
plt.rcParams["figure.figsize"] = (16, 10)
plt.plot(time, speed)
plt.xlabel("Time, hour", fontsize=15, color="g")
plt.ylabel("Speed km/h", fontsize=15, color="g")
plt.title("Значення точок швидкості", fontsize=20, color='g')
plt.grid()
```

- Виконайте інтерполяцію за допомогою interp1d( kind='cubic') та отримайте функцію на 10000 значення. Побудуйте безперервний графік отриманої функції.

```python
f = interp1d(time, speed, kind='cubic')
extra_time = np.linspace(0, 11, 10000)
plt.rcParams["figure.figsize"] = (16, 10)
plt.plot(extra_time, f(extra_time))
plt.xlabel("Time, hour", fontsize=15, color="g")
plt.ylabel("Speed km/h", fontsize=15, color="g")
plt.title("Безперервний графік отриманої функції", fontsize=20, color='g')
plt.grid()
```

- Обчисліть інтеграл для отриманої інтерполяційної функції на проміжку [0, 11]

```python
distance_cubic = integrate.quad(f, 0, 11)
distance_cubic
```

- Виконайте також пункти 4 та 5 для kind='quadratic'

```python
f = interp1d(time, speed, kind='quadratic')
extra_time = np.linspace(0, 11, 10000)
plt.rcParams["figure.figsize"] = (16, 10)
plt.plot(extra_time, f(extra_time))
plt.xlabel("Time, hour", fontsize=15, color="g")
plt.ylabel("Speed km/h", fontsize=15, color="g")
plt.title("Безперервний графік отриманої функції", fontsize=20, color='g')
plt.grid()
```

```python
distance_quadratic = integrate.quad(f, 0, 11)
distance_quadratic
```
